### Data visualization Assignment

**Problem:** Transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data
point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
fig = plt.figure(1, figsize=(8,6))
ax = Axes3D(fig, elev=48, azim=134)

pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)

ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.rainbow_r,edgecolor='k')
ax.set_title("First 3 PCA directions")
ax.set_xlabel("1st Eigen vector")
ax.set_ylabel("2nd Eigen vector")
ax.set_zlabel("3rd Eigen vector")
```
